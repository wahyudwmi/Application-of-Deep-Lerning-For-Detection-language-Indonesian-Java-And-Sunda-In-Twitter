```
def prediksiku(data):
  jumlah_data = len(data)
  seq = tokenizer.texts_to_sequences(data)
  padded = pad_sequences(seq, maxlen=max_length)
  pred= history_blstm.predict(padded)
  pred1 = history_cnn.predict(padded)
  pred2 = history_blstmc.predict(padded)
  pred3 = history_gru.predict(padded)
  pred4 = history_gruc.predict(padded)
  labels = ['Indonesia','Jawa', 'Sunda']

  for z in range(jumlah_data):
      print(data[z], '\n -Teridentifikasi sebagai bahasa : ')
      print('  \033[4mModel   |  Bahasa\033[0m  ')
      print('  BLSTM   : ',labels[np.argmax(pred[z])], pred[z])
      print('  CNN     : ',labels[np.argmax(pred1[z])], pred1[z])
      print('  BLSTM-C : ',labels[np.argmax(pred2[z])], pred2[z])
      print('  GRU     : ',labels[np.argmax(pred3[z])], pred3[z])
      print('  GRU-C   : ',labels[np.argmax(pred4[z])], pred4[z], '\n\n')
```

```

def prediksi_blstm(str1):
  str1 = [str1]
  seq = tokenizer.texts_to_sequences(str1)
  padded = pad_sequences(seq, maxlen=max_length)
  pred = history_blstm.predict(padded)
  labels = ['Indo','Jawa','Sunda']
  str1 = labels[np.argmax(pred)]
  return str1

def prediksi_cnn(str1):
  str1 = [str1]
  seq = tokenizer.texts_to_sequences(str1)
  padded = pad_sequences(seq, maxlen=max_length)
  pred = history_cnn.predict(padded)
  labels = ['Indo','Jawa','Sunda']
  str1 = labels[np.argmax(pred)]
  return str1

def prediksi_blstmc(str1):
  str1 = [str1]
  seq = tokenizer.texts_to_sequences(str1)
  padded = pad_sequences(seq, maxlen=max_length)
  pred = history_blstmc.predict(padded)
  labels = ['Indo','Jawa','Sunda']
  str1 = labels[np.argmax(pred)]
  return str1

def prediksi_gru(str1):
  str1 = [str1]
  seq = tokenizer.texts_to_sequences(str1)
  padded = pad_sequences(seq, maxlen=max_length)
  pred = history_gru.predict(padded)
  labels = ['Indo','Jawa','Sunda']
  str1 = labels[np.argmax(pred)]
  return str1

def prediksi_gruc(str1):
  str1 = [str1]
  seq = tokenizer.texts_to_sequences(str1)
  padded = pad_sequences(seq, maxlen=max_length)
  pred = history_gruc.predict(padded)
  labels = ['Indo','Jawa','Sunda']
  str1 = labels[np.argmax(pred)]
  return str1
```

```
def try_model():
  df0 = pd.read_excel('/content/drive/Shared drives/TA-NLP/WAHYU DWI AMIRULLAH - 15523222/DATASET/@Moektito_tweet-clean.xlsx')
  df0.Text = df0.Text.astype(str)
  df0['blstm'] = df0['Text'].apply(prediksi_blstm)
  df0['cnn'] = df0['Text'].apply(prediksi_cnn)
  df0['blstmc'] = df0['Text'].apply(prediksi_blstmc)
  df0['gru'] = df0['Text'].apply(prediksi_gru)
  df0['gruc'] = df0['Text'].apply(prediksi_gruc)

  df0.assign(Text=df0['blstm'])
  df0.assign(Text1=df0['cnn'])
  df0.assign(Text2=df0['blstmc'])
  df0.assign(Text3=df0['gru'])
  df0.assign(Text4=df0['gruc'])

  writer = pd.ExcelWriter('/content/drive/Shared drives/TA-NLP/WAHYU DWI AMIRULLAH - 15523222/DATASET/@Moektito_tweet-clean skenario 5.xlsx')
  df0.to_excel(writer, 'Sheet1')
  writer.save()
```
